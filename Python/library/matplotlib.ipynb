```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 曲线图
x = np.linspace(-np.pi, np.pi, num=256, endpoint=True)
c, s = np.cos(x), np.sin(x)
plt.plot(x, c, label='$cos(x)$')
plt.plot(x, s, label='$sin(x)$')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

```python
# set figure size and dpi
plt.figure(figsize=(6, 4), dpi=80)

# set subplot
plt.subplot(1, 1, 1)

x = np.linspace(-np.pi, np.pi, num=256, endpoint=True)
c, s = np.cos(x), np.sin(x)

# plot sin(x)
plt.plot(x, s, color='blue', linewidth=1.0, linestyle='-', label='$\sin{x}$')
# plot cos(x)
plt.plot(x, c, color='red', linewidth=1.0, linestyle='-', label='$\sin{x}$')

# 设置坐标轴上的点
# plt.xticks(np.linspace(-4, 4, 9, endpoint=True))
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
            [r'$-\pi$', r'$-\frac{\pi}{2}$', r'$0$', r'$\frac{\pi}{2}$', r'$+\pi$'])
plt.yticks(np.linspace(-1.0, 1.0, 3, endpoint=True))

# 设置坐标轴范围
# plt.xlim(-4.0, 4.0)
# plt.ylim(-1.2, 1.2)
xmin ,xmax = x.min(), x.max()
ymin, ymax = c.min(), c.max()

dx = (xmax - xmin) * 0.1
dy = (ymax - ymin) * 0.1

plt.xlim(xmin - dx, xmax + dx)
plt.ylim(ymin - dy, ymax + dy)

# set legend
plt.legend(loc='upper left')

# set axis to center
# 共有四条坐标轴，上下左右
ax = plt.gca()
ax.spines['right'].set_color('none') # left set none color
ax.spines['top'].set_color('none') 
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))

# set some mark on figure
t = 2*np.pi/3
plt.plot([t,t], [0,np.cos(t)], color ='red', linewidth=2.5, linestyle="--")
plt.scatter([t,], [np.cos(t),], color ='red')

plt.annotate(r'$\sin(\frac{2\pi}{3})=\frac{\sqrt{3}}{2}$',
         xy=(t, np.sin(t)), xycoords='data',
         xytext=(+10, +30), textcoords='offset points', fontsize=13,
         arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))      # 设置标记

plt.plot([t,t],[0,np.sin(t)], color ='blue', linewidth=2.5, linestyle="--")
plt.scatter([t,],[np.sin(t),], 50, color ='blue')

plt.annotate(r'$\cos(\frac{2\pi}{3})=-\frac{1}{2}$',
         xy=(t, np.cos(t)), xycoords='data',
         xytext=(-90, -50), textcoords='offset points', fontsize=13,
         arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

# plot
plt.show()

# save
# plt.save('')
```

```python
# scatter figure

# set figure size
plt.figure(figsize=(5, 5), dpi=80)

# set subplot
plt.subplot(1, 1, 1)

# set lim
plt.xlim(-1.0, 1.0)
plt.ylim(-1.0, 1.0)

# set axis
ax = plt.gca()
ax.spines['right'].set_color('none') # left set none color
ax.spines['top'].set_color('none') 
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))

# set ticks
plt.xticks([-1.0, -0.5, 0, 0.5, 1.0])
plt.yticks([-1.0, -0.5, 0.5, 1.0])

# make data
SqrtTwoByTwo = np.sqrt(2)/2
x = np.ones(20) * SqrtTwoByTwo
x = np.append(x, np.ones(20) * (-SqrtTwoByTwo))
x = np.append(x, np.ones(20) * SqrtTwoByTwo)
x = np.append(x, np.ones(20) * (-SqrtTwoByTwo))
y = np.ones(40) * SqrtTwoByTwo
y = np.append(y, np.ones(40) * (-SqrtTwoByTwo))
# add noise
SNR = 8
CNR = SNR + 3.0
sigma2 = 10 ** ((-CNR)/10)
x = x + np.random.randn(80) * sigma2
y = y + np.random.randn(80) * sigma2

plt.scatter(x, y, s=10, c='blue')
plt.scatter([SqrtTwoByTwo, -SqrtTwoByTwo, -SqrtTwoByTwo, SqrtTwoByTwo], 
        [SqrtTwoByTwo, SqrtTwoByTwo, -SqrtTwoByTwo, -SqrtTwoByTwo], s=40, c='red', marker='*')

plt.show()
```
